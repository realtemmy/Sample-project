## 0. An end to end Scikit-Learn workflow

```python
# Get the data ready
import pandas as pd
import numpy as np
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
# Create  X(features matrix)
x = heart_disease.drop("target", axis=1)

# Create y (labels)
y = heart_disease["target"]
```

```python
# Choose theright model and hyperparameters
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

# We'll keepthe default hyperparameters
clf.get_params()
```

```python
# 3. Fit the model to the training data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
clf.fit(x_train, y_train);
```

```python
x_train
```

```python
x_test
```

```python
y_preds = clf.predict(x_test)
y_preds
```

```python
y_test
```

```python
# 4. Evaluate the model
clf.score(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

```python
# %. Improve a model
# Try different amount of n_estimators
np.random.seed(42)
for i in range(10, 100, 10):
    print(f"Trying model {i} estimators...")
    clf = RandomForestClassifier(n_estimators=i).fit(x_train, y_train)
    print(f"Model accuracy on test set: {clf.score(x_test, y_test) * 100 }.2f%")
    print("")
```

```python
# 6. Ssave a model and load it
import pickle
pickle.dump(clf, open("random_first_model_1.pk1", "wb"))
```

```python
loaded_model = pickle.load(open("random_first_model_1.pk1", "rb"))
loaded_model.score(x_test, y_test)
```

## Split into X and Y

```python
x = car_sales.drop("Price", axis=1)
y = car_sales["Price"]


# Split into training data set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
# Turn the categories into numbers
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
categorical_features = ["Make", "Colour", "Doors"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot", one_hot, categorical_features)], remainder="passthrough")

transformed_X = transformer.fit_transform(x)
transformed_X
```

```python
pd.DataFrame(transformed_X)
```

```python
# Another way to convert all data to numerical
dummies = pd.get_dummies(car_sales[["Make", "Colour", "Doors"]])
dummies
```

```python
# Let's refit the model
np.random.seed(42)
x_train, x_test, y_train, y_test = train_test_split(transformed_X, y, test_size=0.2)
model.fit(x_train, y_train)
```
