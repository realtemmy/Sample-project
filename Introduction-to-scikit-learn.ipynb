### Fill missing values with Scikit-learn

```python
car_sales_missing = pd.read_csv("car-sales-extended-missing-data.csv");
car_sales_missing.head()
```

```python
car_sales_missing.isna().sum()
```

```python
# Drop the rows with no labels
car_sales_missing.dropna(subset=["Price"], inplace=True);
car_sales_missing.isna().sum()
```

```python
# Split into x and y
x = car_sales_missing.drop("Price", axis=1);
y = car_sales_missing["Price"]
```

```python
# Fill missing values with Scikit-Learn
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

# Fill categorical values with 'missing' and numerical values with mean
cat_inputer = SimpleImputer(strategy="constant", fill_value="missing");
door_imputer = SimpleImputer(strategy="constant", fill_value=4);
num_imputer = SimpleImputer(strategy="mean");

# Define columns
cat_features = ["Make", "Colour"]
door_features = ["Doors"]
num_features = ["Odometer (KM)"]


imputer = ColumnTransformer([
    ("cat_imputer", cat_inputer, cat_features),
    ("door_imputer", door_imputer, door_features),
     ("num_imputer", num_imputer, num_features),
])

# Transform the data
filled_X = imputer.fit_transform(x)
filled_X
```

```python
car_sales_filled = pd.DataFrame(filled_X, columns=["Make", "Colour", "Doors", "Odometer (KM)"])
car_sales_filled.head()
```

```python
car_sales_filled.isna().sum()
```

```python
# Convert data to numerical(numbers)
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = ["Make", "Colour", "Doors"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot", one_hot, categorical_features)], remainder="passthrough")
transformed_X = transformer.fit_transform(car_sales_filled)
transformed_X
```

```python
# Now that data is represented as numbers and no missing values
# Let's fit a model
np.random.seed(42)

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(transformed_X, y, test_size=0.2);
model = RandomForestRegressor()
model.fit(x_train, y_train);
model.score(x_test, y_test)
```

# 2.2 Picking a machine learning model for a classification problem

```python
# Using breast cancer datasets
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
data
```

```python
data_df = pd.DataFrame(data["data"], columns=data["feature_names"])
data_df
```

```python
data_df["target"] = data["target"]
data_df
```

```python
# Using classification (SDGClassifier)
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import train_test_split

x = data_df.drop("target", axis=1)
y = data_df["target"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

clf = SGDClassifier(loss="hinge", penalty="l2", max_iter=1000)
clf.fit(x_train, y_train)
clf.score(x_test, y_test)
```

```python
# Classification Kernel approximation
from sklearn.kernel_approximation import RBFSampler
from sklearn.linear_model import SGDClassifier

# X and Y values
x = data_df.drop("target", axis=1)
y = data_df["target"]

rbf_feature = RBFSampler(gamma=1, random_state=1)
x_features = rbf_feature.fit_transform(x)
clf = SGDClassifier(max_iter=1000)
clf.fit(x_features, y)
clf.score(x_features, y)
```

```python
# Classification (Ensemble classifiers)
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.datasets import make_hastie_10_2

x = data_df.drop("target", axis=1)
y =  data_df["target"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

clf = HistGradientBoostingClassifier()
clf.fit(x_train, y_train)
clf.score(x_test, y_test)
```

```python
# Classification, using RandonForestClassifier in ensemble
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_hastie_10_2

x = data_df.drop("target", axis=1)
y =  data_df["target"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

clf = RandomForestClassifier()
clf.fit(x_train, y_train)
clf.score(x_test, y_test)
```

Tidbit:
    1. If there's structured data, use ensemble method
    2. If there's unstructured data eg audio, text files etc, use deep learning or transfer learning
